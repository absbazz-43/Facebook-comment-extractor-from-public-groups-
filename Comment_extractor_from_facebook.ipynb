## Function for facebook comment extractor

```python
import pandas as pd
import facebook_scraper as fs

def comment_extractor(no_comments, post_id):
    gen = fs.get_posts(
        post_id,
        options = {'comments': no_comments, "progress": True}
    )
    post = next(gen)
    comments = post['comments_full']
    coment = []
    for comment in comments:
        coment.append(comment)
    comment_data = pd.DataFrame(coment)
    return comment_data
```

```python
comment_extractor(100,"299184380613465")
```

## Function for post extractor

```python
import facebook_scraper as fs

def post_comment_extractor(POSTID, no_comments):
    gen = fs.get_posts(
    post_urls = [POSTID],
    options = {"comments": no_comments, "progress": True}
    )
    post = next(gen)
    comments = post['comments_full']
    commments = []
    for comment in comments:
        commments.append(comment)
    extract_data = pd.DataFrame(commments)
    return extract_data
```
