# Language and transation

```python

###
def translator_bn(text):
  lan = detect(text)
  if lan =="bn":
    return text
  else :
    txt_spliter = text.split()
    actual_text =[]
    stop_words = set(stopwords.words('english'))
    for w in txt_spliter:
      if w not in stop_words:
        actual_text.append(w)
    bn = []
    for i in actual_text:
      bn.append(GoogleTranslator(source='en', target='bn').translate(text=i))
    return bn

#return txt_spliter
text = "amar product ta hariye giyeche"
translator_bn(text)
```
